```python
import pandas
import math

pandas.set_option("display.precision", 12)

def sq_side(sq :float):
    val = math.sqrt(4 - sq)
    return 2 - val

def new_data(data: dict):
    new = {}
    new['lados'] = 2 * data['lados']
    new['lado_cuadrado'] = sq_side(data['lado_cuadrado'])
    new['lado'] = math.sqrt(new['lado_cuadrado'])
    new['perímetro'] = new['lados'] * new['lado']
    new['semiperímetro'] = new['perímetro'] / 2
    new['área_triángulo'] = new['lado'] * (1 - sq_side(new['lado_cuadrado']) / 2) / 2
    new['área_total'] = new['lados'] * new['área_triángulo']
    return new

first_area = 1 * (1 - sq_side(1) / 2) / 2

first_data = {'lados': 6,
              'lado_cuadrado': 1,
              'lado': 1,
              'perímetro': 6,
              'semiperímetro': 3,
              'área_triángulo': first_area,
              'área_total': 6 * first_area
              }

table = [first_data]
```

```python
for i in range(10):
    table.append(new_data(table[-1]))
```

```python
data = pandas.DataFrame.from_dict(table)

data.head(11)
```

```python
data.to_csv('geo_circle.csv')
```
